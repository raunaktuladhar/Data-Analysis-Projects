# Handling Missing Values

```python
df['last review'] = pd.to_datetime(df['last review'], errors = 'coerce')
# coerce -> if any missing date, then it will convert to NAT (Not A Thing)
```

```python
df.info()
```

```python
df.fillna({'reviews per month' : 0, 'last review' : df['last review'].min()}, inplace = True)
# inplace = True means permanent change 
```

```python
df.dropna(subset = ['NAME', 'host name'], inplace=True)
```

```python
print(df.isnull().sum())
```

```python
df = df.drop(columns=['license', 'house_rules'], errors = 'ignore')
```

```python
df.head()
```

```python
# remove $ and convert into float
df['price'] = df['price'].replace('[\$,]', '', regex=True).astype(float)
df['service fee'] = df['service fee'].replace('[\$,]', '', regex=True).astype(float)
```

```python
df.head()
```

# remove duplicate values

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

```python
# descriptive stats
df.describe()
```

## Task no. 1

```python
plt.figure(figsize= (10,6))  # sets the width and height of the figure in inches; in this line of code, width=10 and height=6
sns.histplot(df['price'], bins=50, kde=True, color='red') # kde controls curve in graph
plt.title("Distribution of Listing price")
plt.xlabel("Price in $")
plt.ylabel("Frequency")
plt.show()  # avoids showing Text(0, 0.5, 'Frequency'); () should also be there
```

## Conclusion:
The histogram shows a fairly even distribution of listing prices across different price ranges, indicating no particular concentration of listings in any specific price range. The KDE line helps to visualize this even spread more clearly, confirming that the dataset contains listings with a wide variety of prices

## Task no 2

```python
df['room type']
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='room type', data=df, color='hotpink')
plt.title("Room Type Distribution")
plt.xlabel("Room Type")
plt.ylabel("Count")
plt.show()
```

## Conclusion:
Room types are distributed by the following:

## Task no 4

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='room type', y='price', hue='room type', data=df, palette='Set1')
plt.title("Room Type vs Price")
plt.xlabel("Room Type")
plt.ylabel("Price in $")
plt.legend(title='Room Type')
plt.show()
```

## Conclusion:
1. Private room costs around 400 - 900
2. Entire home / Apartment costs around same as Private room
3. Shared room costs around same as Private room and Apartment
4. Hotel room costs around 500 - 900

## Task no 5

```python
# last review attribute is converted to datetime format in terms of month
df['last review'] = pd.to_datetime(df['last review'])
reviews_over_time = df.groupby(df['last review'].dt.to_period('M')).size()

plt.figure(figsize=(12,6))
reviews_over_time.plot(kind='line', color='red')
plt.title('No. of Reviews over time')
plt.xlabel('Date')
plt.ylabel('No. of Reviews')
plt.show()
```

## Conclusion:
The year 2019 had the highest no. of reviews
